### Combined Dataframes using Merge

Just to make sure that we don't have unrelated (orphan data) we can merge the dataframes to make one combined dataframe to facilitate analysis.

````python
merged_movie_lens = rating.merge(user, how='left', left_on='user id', right_on='user id').merge(movie, how='left', left_on='movie id', right_on='movie id')
merged_rating_user = rating.merge(user, how='left', left_on='user id', right_on='user id')
merged_rating_movie = rating.merge(movie, how='left', left_on='movie id', right_on='movie id')
print(f'Shape [All 3 - Left Join]: {merged_movie_lens.shape}')
print(f'Shape [Rating -> User - Left Join]: {merged_rating_user.shape}')
print(f'Shape [Rating -> Movie - Left Join]: {merged_rating_movie.shape}')
merged_movie_lens.head()
````

### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use item dataset to get count of movies released in a particular year, store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.

````python
# I Dropped the value of unknown from the columns and also removed the year 971 because it is invalid value.
# I had alread calculated the total movies released in a year and added to column total_genres
# the sum of all genres which is nothing but total number of movies released in a given year
# because when grouped it is total number of observations
plt.pyplot.figure(figsize=(20,20)) 
release_by_year = movie.groupby('release year').sum().drop(columns=['movie id', 'unknown']).drop(labels=['971'])
release_by_year_percentage = release_by_year.div(release_by_year['total_genres'], axis=0)*100
sns.heatmap(release_by_year_percentage.drop(axis=1, columns=['total_genres']))
````

**Insights:**

- Drama is an evergreen genre in terms of movie production and is being constantly favoured by film makers
- Western genre is on constant decline and has almost gone missing after 1990s
- Comedy movies are also very abundant.
- Quite a bit of movies made in 1976 were horror movies.
- Almost all the movies produced in 1967 were drama movies.

### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

````python
ratings_by_title = merged_rating_movie.drop(columns=['user id', 'movie id', 'timestamp', 'release date', 'unknown',
'Fantasy',
'Film-Noir',
'Western',
'Animation',
'Documentary',
'Musical',
'Mystery',
'War',
'Horror',
'Sci-Fi',
'Crime',
'Childrens',
'Adventure',
'Romance',
'Action',
'Thriller',
'Comedy',
'Drama']).groupby('movie title').agg(['count', 'mean'])
ratings_by_title.columns = ["_".join(x) for x in ratings_by_title.columns.ravel()]
ratings_by_title[ratings_by_title.rating_count>100].sort_values(by='rating_mean',ascending=False).head(25)
````
