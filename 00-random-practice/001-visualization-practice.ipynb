```python
import pandas as pd
import seaborn as sns   # Why sns?  It's a reference to The West Wing
import matplotlib.pyplot as plt  # seaborn is based on matplotlib
sns.set(color_codes=True) # adds a nice background to the graphs
%matplotlib inline 
# tells python to actually display the graphs
```

```python
auto = pd.read_csv('Automobile.csv')
auto.head()
```

```python
sns.set({'figure.figsize': (8,6)})
sns.histplot(auto['highway_mpg'], kde=True);
```

```python
# we can turn the kde off and put a tic mark along the x-axis for every data point with rug
sns.set({'figure.figsize': (8,6)})
sns.histplot(auto['city_mpg'], kde=False);
```

```python
sns.set({'figure.figsize': (8,8)})
sns.jointplot(x='engine_size', y='horsepower', data=auto);
```

```python
sns.jointplot(x='engine_size', y='horsepower', data=auto, kind="hex");
```

```python
sns.jointplot(x='engine_size', y='horsepower', data=auto, kind="kde");
```

```python
sns.pairplot(auto[['normalized_losses', 'engine_size', 'horsepower', 'highway_mpg', 'city_mpg']]);
```

```python
sns.stripplot(x='fuel_type', y='horsepower', data=auto, jitter=True);
```

```python
sns.swarmplot('fuel_type', 'horsepower', data=auto);
```

```python
sns.boxplot(x='number_of_doors', y='horsepower', data=auto);
```

```python
sns.boxplot(x='number_of_doors', y='horsepower', data=auto, hue=auto['fuel_type']);
```

```python
sns.barplot(x='body_style', y='horsepower', data=auto, hue=auto['fuel_type']);
```

```python
sns.countplot(x='body_style', data=auto, hue=auto['fuel_type']);
```

```python
sns.pointplot(x='body_style', y='horsepower', data=auto, hue=auto['number_of_doors']);
```

```python
sns.catplot(x="fuel_type",
               y = "horsepower",
               hue="number_of_doors", 
               col="drive_wheels", 
               data=auto, 
               kind="box");
```

```python
sns.lmplot(y="horsepower", x="engine_size", data=auto);
```

```python
sns.lmplot(y="horsepower", x="engine_size",hue="fuel_type", data=auto);
```

```python
sns.pointplot(auto['yr'],auto['acc']); plt.show()
```
