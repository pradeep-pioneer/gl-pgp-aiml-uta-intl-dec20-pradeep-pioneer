# Project - Bank Churn Prediction

## Description

### Background and Context

Businesses like banks that provide service have to worry about the problem of 'Churn' i.e. customers leaving and joining another service provider. It is important to understand which aspects of the service influence a customer's decision in this regard. Management can concentrate efforts on the improvement of service, keeping in mind these priorities.

### Objective

Given a Bank customer, build a neural network-based classifier that can determine whether they will leave or not in the next 6 months.

### Data Description

The case study is from an open-source dataset from Kaggle. The dataset contains 10,000 sample points with 14 distinct features such as CustomerId, CreditScore, Geography, Gender, Age, Tenure, Balance, etc.

#### Data Dictionary

* **CustomerId:** Unique ID which is assigned to each customer
* **Surname:** Last name of the customer 
* **CreditScore:** It defines the credit history of the customer.  
* **Geography:** A customerâ€™s location    
* **Gender:** It defines the Gender of the customer   
* **Age:** Age of the customer     
* **Tenure:** Number of years for which the customer has been with the bank
* **NumOfProducts:** It refers to the number of products that a customer has purchased through the bank.
* **Balance:** Account balance
* **HasCrCard:** It is a categorical variable that decides whether the customer has a credit card or not.
* **EstimatedSalary:** Estimated salary 
* **isActiveMember:** It is a categorical variable that decides whether the customer is an active member of the bank or not ( Active member in the sense, using bank products regularly, making transactions, etc )
* **Excited:** It is a categorical variable that decides whether the customer left the bank within six months or not. It can take two values.
```
                    0=No ( Customer did not leave the bank )
                    1=Yes ( Customer left the bank )
```

# Data Pre-processing

- Split the target variable and predictors
- Split the data into train and test
- Rescale the data

## Split the target variable and predictors

````python
X_bank_customers = df_bank_customers_cleaned.drop('Exited', axis=1)
y_bank_customers = df_bank_customers_cleaned['Exited']
X_bank_customers = pd.get_dummies(X_bank_customers, drop_first=True)
````

## Split the data into train and test

### Pre-requisites
Before being able to split the data - we need to treat the categorical variables like ```Gender``` and ```Geography```.

````python
X_train, X_test, y_train, y_test = train_test_split(X_bank_customers, y_bank_customers, test_size = 0.2, random_state = 0, stratify = y_bank_customers)
X_train.head()
````

## Rescale the data
We will fit the standard scaler on training data and apply the transformations to test data and training data.

````python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train
````

# Model building

We will build a fully connected network with 3 layers, the first input layer will be a ```Dense``` layer with neurons equal to features in our in our training data (independent variables), second layer will be another ```Dense``` layer 128 neurons. In the output layer (again ```Dense``` layer) we will have only one nueron with **Sigmoid** as the activation function because we only have 2 output classes.

Here is a layout of our Neural Network:
1. Input Layer:
  1. Type: ```Dense```
  2. Neurons: 11
  3. Activation: ```relu```
2. First Hidden Layer:
  1. Type: ```Dense```
  2. Neurons: 128
  3. Activation: ```relu```
3. Output Layer:
  1. Type: ```Dense```
  2. Neurons: 1
  3. Activation: ```sigmoid```

In addition we will use the following parameters for the Network.

- **Optimizer**: ```adam```
- **Loss**: ```BinaryCrossentropy```

We will create a helper function to help us recompile the model several times with different parameters.
We will care for the following *metrics* while building our model.
* **Accuracy**
* **Precision**
* **Recall**
* **TruePositives**
* **FalsePositives**
* **TrueNegatives**
* **FalseNegatives**
* **AUC**
* **Precision Recall Curve (PRC)**

````python
model_metrics = [
      keras.metrics.MeanSquaredError(name='mse'),
      keras.metrics.TruePositives(name='t_pos'),
      keras.metrics.FalsePositives(name='f_pos'),
      keras.metrics.TrueNegatives(name='t_neg'),
      keras.metrics.FalseNegatives(name='f_neg'), 
      keras.metrics.BinaryAccuracy(name='accuracy'),
      keras.metrics.Precision(name='precision'),
      keras.metrics.Recall(name='recall'),
      keras.metrics.AUC(name='auc'),
      keras.metrics.AUC(name='prc', curve='PR')
]
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_prc', 
    verbose=1,
    patience=10,
    mode='max',
    restore_best_weights=True)
````

````python
def create_model(metrics=model_metrics, optimizer = keras.optimizers.Adam(learning_rate=0.001), loss=keras.losses.BinaryCrossentropy(), classifier_bias=None):
  if classifier_bias is not None:
    classifier_bias = tf.keras.initializers.Constant(classifier_bias)
  model = Sequential()
  model.add(Dense(X_train.shape[1], activation='relu', input_dim = X_train.shape[1]))
  model.add(Dense(128, activation='relu'))
  model.add(Dense(1, activation='sigmoid', bias_initializer=classifier_bias))
  model.compile(
      optimizer=optimizer,
      loss=loss,
      metrics=metrics)

  return model
````

````python
model = create_model()
training_history_1 = model.fit(X_train, y_train, epochs=30, callbacks=[early_stopping], validation_split = 0.2, batch_size=256, verbose=True)
````

````python
def see_loss(history, label, n):
  # Use a log scale on y-axis to show the wide range of values.
  plt.semilogy(history.epoch, history.history['loss'],
               label='Train ' + label)
  plt.semilogy(history.epoch, history.history['val_loss'],
               label='Val ' + label,
               linestyle="--")
  plt.xlabel('Epoch')
  plt.ylabel('Loss')
````

````python
def see_mse(history, label, n):
  # Use a log scale on y-axis to show the wide range of values.
  plt.semilogy(history.epoch, history.history['mse'],
               label='Train ' + label)
  plt.semilogy(history.epoch, history.history['val_mse'],
               label='Val ' + label,
               linestyle="--")
  plt.xlabel('Epoch')
  plt.ylabel('MSE')
````

````python
see_loss(training_history_1, "Loss", 0)
````

````python
see_mse(training_history_1, "MSE", 0)
````

````python
def create_confusion_matrix(truth, predicted):
  cm = confusion_matrix(truth, predicted > 0.5)
  plt.figure(figsize=(5,5))
  sns.heatmap(cm, annot=True, fmt="d")
  plt.title('Confusion matrix @{:.2f}'.format(0.5))
  plt.ylabel('Actual label')
  plt.xlabel('Predicted label')
  print('True Negatives: ', cm[0][0])
  print('False Positives: ', cm[0][1])
  print('False Negatives: ', cm[1][0])
  print('True Positives: ', cm[1][1])
  print('Total Positives: ', np.sum(cm[1]))
````

````python
y_predict = model.predict(X_test)
create_confusion_matrix(y_test, y_predict)
````

# Model Performance Improvement

## How is the current model performing? (Comment on model performance)
The current model performance is not very good - two of the three key matrices that we care about are not having their best values - which takes us to the next question.

## Comment on which metric is right for model performance evaluation and why?

On a very high level we care about three key matrices:
* Accuracy: Informally, accuracy is the fraction of predictions our model got right, It is formalized as Correct Predictions / All Predictions.
* Precision: It helps in answering - What proportion of positive identifications was actually correct? It is calculated as True Positives / True Positives + False Positives
* Recall: What proportion of actual positives was identified correctly? It is calculated as True Positives/True Positives + False Negatives.

**Why do we care about these metrics?**
Our goal is to be able to predict the potential churn customers which the bank might end up loosing, while accuracy gives us an indication as to how many predictions the model is able to make are near truth - it doesn't indicate the success rate for the classes that we care about the most. More specifically accuracy is not good indicator if we want our model to be really good in predicting one specific class to maximum possible correctness.

Understanding the business implications of False Positives & False Negatives are listed below:
1. False Positives: If our model produces a lot of False Positives then we waste resources on customers who actually wouldn't have churned (same could have been used on other customer who actually churned).
2. False Negatives: If your model generates a lot of False Negatives then we would loose customers leading to revenue loss.

Now with this and the above definition we can see that we would want to have a good precision & recall so that our model is able to tell us the maximum potential customers who might churn.


## Checking AUC & PRC (precision-recall curve)

````python
def plot_metrics(history):
  metrics = ['prc', 'auc']
  for n, metric in enumerate(metrics):
    name = metric.replace("_"," ").capitalize()
    plt.subplot(2,2,n+1)
    plt.plot(history.epoch, history.history[metric], label='Train')
    plt.plot(history.epoch, history.history['val_'+metric], linestyle="--", label='Val')
    plt.xlabel('Epoch')
    plt.ylabel(name)
    if metric == 'auc':
      plt.ylim([0.8,1])
    else:
      plt.ylim([0,1])

    plt.legend()
````

````python
plot_metrics(training_history_1)
````

````python
def plot_prc(name, labels, predictions):
    precision, recall, _ = sklearn.metrics.precision_recall_curve(labels, predictions)

    plt.plot(precision, recall, label=name, linewidth=2)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.grid(True)
    ax = plt.gca()
    ax.set_aspect('equal')
````

````python
plot_prc("Test Baseline", y_test, y_predict)
````

## Can model performance be improved?

The model performance is suffering because we have imbalanced dataset, there are a few things we can play with to see if we can get better results. We will try the following:
1. Changing the Classweights: We can configure the class weights so that the model can pay special attention to the minority class and adapt so that we don't have a model which is biased toward the majority class.
2. Tuning the bias for output layer: We can tweak the bias for output layer so that it favors the class that we care about rather than the majority class.
3. Oversampling the data: We can oversample the data using the ```tf.data``` module available in tensorflow.

Let's start with first option.

### Class weights
#### Calculate class weights
We can calculate the classweights by ```(1 / class observations) * (total / 2.0)``` for each class.

````python
neg, pos = np.bincount(df_bank_customers_cleaned['Exited'])
print(f'Negatives: {neg}')
print(f'Positives: {pos}')
````

````python
weight_for_neg = (1 / neg) * (neg + pos / 2.0)
weight_for_pos = (1 / pos) * (neg + pos / 2.0)
output_weights = {0: weight_for_neg, 1: weight_for_pos}
print(f'Output weights: {output_weights}')
weighted_model = create_model()
training_history_2 = weighted_model.fit(
    X_train, 
    y_train, 
    epochs=30, 
    callbacks=[early_stopping], 
    validation_split = 0.2, 
    batch_size=256,
    class_weight=output_weights,
    verbose=True)
````

````python
see_loss(training_history_2, "Loss", 0)
````

#### Outcome
There is an improvement on model performance in terms of recall after adjusting the class weights but now the precision has dropped significantly. This means that we have traded precison for recall. All the three metrics that we care for still remain quite low.

### Tuning the bias for output layer:
We will now tune the bias for the output layer to try and tune the outcome from the output layer, we can use the following formula to calculate the values for the bias: ```log(pos/neg)```.
